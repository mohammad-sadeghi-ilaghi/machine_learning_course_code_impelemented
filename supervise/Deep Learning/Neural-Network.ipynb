```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_moons, make_circles, make_classification
```

```python
# X, y = make_moons(n_samples=3000, noise=0.15)
# X, y = make_circles(n_samples=3000, noise=0.14, factor=0.3)
X, y = make_classification(n_samples=3000, n_features=2, n_redundant=0)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
from keras import Sequential
from keras.layers import  Dense
```

```python
model = Sequential()
model.add(Dense(10, activation='relu', input_shape= (2,) ) ) # input shape is given for the first hidden layer.
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', metrics='accuracy')
```

```python
model.fit(X, y, batch_size= 64, epochs=50, validation_split=0.1)
```
